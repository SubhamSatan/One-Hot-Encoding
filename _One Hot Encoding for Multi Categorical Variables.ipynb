```python
#lets examine how many columns we will obtain after one hot encoding these variable
pd.get_dummies(data,drop_first=True).shape
```

```python
# lets find the top most frequent categories for the variable X2
top = data.X2.value_counts().sort_values(ascending=False).head(10).index
top
```

```python
# top 10 categoreis of column X2
top_10 = [x for x in data.X2.value_counts().sort_values(ascending=False).head(10).index]
```

```python
top_10
```

```python
# for label we make the 10 binary variables 

for label in top_10:
    data[label]=np.where(data['X2']==label,1,0);
    
data[['X2']+top_10].head(40)    
    
```
